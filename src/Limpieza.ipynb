Para la realización de la base de datos, se van a concatenar las tablas de saque, resto y ranking, ya que tienen el elemnto jugador en común y todas tienen la misma longitud.

Para ello se ordenará de manera alfabética la columna de jugador y se concatenerán las tablas. La tabla inicial será la tabla de ranking.

```python
ranking = ranking.sort_values(by='Jugador')
ranking.reset_index(inplace=True)
ranking = ranking.drop(columns='index')
```

```python
saque = saque.sort_values(by='Jugador')
saque.reset_index(inplace=True)
saque = saque.drop(columns='index')
```

```python
resto = resto.sort_values(by='Jugador')
resto.reset_index(inplace=True)
resto = resto.drop(columns='index')
```

```python
#Como se van a concatenar las listas, no es necesario tener la columna de nombres en las tres.

#Se elimina la columna de Jugador en resto y saque:

resto = resto.drop(columns='Jugador')
saque = saque.drop(columns='Jugador')
```

```python
#Se unen los tres DF:

jugadores = pd.concat([ranking, saque, resto], axis=1)
```

```python
#Ordenamos de nuevo por nº en el ranking:

jugadores = jugadores.sort_values(by='Posicion_Ranking')
```

```python
jugadores.reset_index(inplace=True)

jugadores = jugadores.drop(columns='index')
```

```python
#Se coloca la columna jugador en primera posición:

first_column = jugadores.pop('Jugador') 
  
jugadores.insert(0, 'Jugador', first_column)
```

Se comprueba que en apellidos compuestos como Auger-aliassime el segundo apellido comienza con minúscula, diferencia de como se muestra en el resto de tablas. Solo ocurre con este apellido por que se hace una función para modificar especificamente este:

```python
def Auger(x):
    
    if 'Aliassime'== x:
        
        return 'Auger-aliassime'
    
    elif x!='Aliassime':
        
        return x
```
